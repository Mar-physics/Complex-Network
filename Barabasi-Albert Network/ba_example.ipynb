```python
import networkx as nx
import matplotlib.pyplot as plt

# Parameters for the Barabási-Albert model
n = 1000  # Number of nodes
m = 1   # Number of edges to attach from a new node to existing nodes

# Generate Barabási-Albert network
BA_network = nx.barabasi_albert_graph(n, m)

# Calculate the degree of each node
degree_dict = dict(BA_network.degree())
max_degree = max(degree_dict.values())

# Set a threshold for hub nodes
threshold = max_degree/10

# Create a list for node colors
node_colors = ['red' if degree_dict[node] >= threshold else 'skyblue' for node in BA_network.nodes()]

# Draw the network with different colors for hubs and non-hubs
plt.figure(figsize=(8, 8))
nx.draw(BA_network, with_labels=False, node_size=50, node_color=node_colors, edge_color="gray", alpha=0.7)

# Show the plot
plt.title("Barabási-Albert Network with Hubs Highlighted")
plt.show()
```

```python
#max degree
print(max_degree)
```

```python
#mean number of links per node

mean_degree = sum(degree_dict.values())/n

print(mean_degree)
```

```python
# Create the adjacency matrix
adjacency_matrix = nx.to_numpy_array(BA_network)

# Set up the plot for the adjacency matrix
plt.figure(figsize=(10, 8))

# Display the adjacency matrix
plt.imshow(adjacency_matrix, cmap='hot_r', interpolation='none')
plt.title('Adjacency Matrix of the Barabási-Albert Network')
plt.xlabel('Node Index')
plt.ylabel('Node Index')


# Show the plot
plt.show()
```
